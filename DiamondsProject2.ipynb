```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
#sns.set_style('darkgrid') #my default is "whitegrid"
```

```python
df = pd.read_csv("diamonds.csv") #1st column needs encoding to read pound/hashtag(#) symbol
df.head()
```

```python
#rename the pound/hashtag(#) column so it doesn't cause further issues
df.rename(columns={'Unnamed: 0': 'Number'}, inplace =True)
```

```python
#validate
df.head()
```

```python
#set indices as Pokemon numbers
df.set_index('Number', inplace=True)
df.head()
```

```python
#make a scatterplot
sns.lmplot(x='price', y='carat', data=df)
```

```python
#add more details to scatterplot
sns.lmplot(x='price', y='carat', data=df, 
           fit_reg=False, #remove regression line
           hue='color')   #color by evolution stage
```

```python
#boxplot showing stats for each characteristic
sns.boxplot(data=df)
```

```python
#create a dataframe containing the stats for each Pokemon
#drop Total, Stage, and Legendary
stats_df = df.drop(['cut', 'clarity', 'color'], axis=1)
stats_df.head()
```

```python
#boxplot for stats
sns.boxplot(data=stats_df)
```

```python
df2=df.head(1000)
```

```python
df2.head()
```

```python
#change size of plot
plt.subplots(figsize=(13,7))
#violin plot shows the distribution of attack power for each type
sns.violinplot(x='price', y='carat', data=df2)
```

```python
#create color palette
#color Hex numbers are from Bulbapedia (https://bulbapedia.bulbagarden.net/wiki/Category:Type_color_templates)
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
#violin plot with Pokemon color palette
plt.subplots(figsize=(13,7))
sns.violinplot(x='price', y='carat', data=df2, palette=pkmn_type_colors)
```

```python
#swarm plot
plt.subplots(figsize=(10,6))
sns.swarmplot(x='price', y='carat', data=df2, palette=pkmn_type_colors)
```

```python
#combine a violin plot with a swarm plot
plt.figure(figsize=(13,7))
sns.violinplot(x='price',
               y='carat',
               data=df2,
               inner=None, #removes bars inside violin plot
               palette=pkmn_type_colors)

sns.swarmplot(x='price',
              y='carat',
              data=df2,
              color='k', #make points black
              alpha=0.7) #slightly transparent

plt.title('Price by Carat')
```
